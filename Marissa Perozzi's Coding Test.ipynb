```python
#load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load data
data=pd.read_csv("LED_CLEAN.csv")
data
#See that variable names 
list(data)
```

```python
#get counts for ethnicity 
data['ethnicity'].value_counts()
```

```python
#pie chart 
labels = ('Black', 'White', 'Asian')
sizes = [170, 383, 51]
colors = ['gold', 'yellowgreen', 'lightcoral']
plt.pie(sizes,labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Sexual Assault Survivors by Race')
plt.show()
```

```python
#counts of assault by age groups
data.groupby(['age', 'are.you.a.survivor.of.sexual.assault?'])['are.you.a.survivor.of.sexual.assault?'].value_counts()
```

```python
#bar plot of age groups
data3 = [154,261,122,50,21,1]
labels = ['18-24', '25-34', '35-44', '45-54','55-64','64-75','75 and Older']
width =1
plt.xticks(range(len(data3)), labels,fontsize=8)
plt.xlabel('Age Groups')
plt.ylabel('Counts')
plt.title('Sexual Assault Victims by Age')
plt.bar(range(len(data3)), data3,color='green') 
plt.show()
```
